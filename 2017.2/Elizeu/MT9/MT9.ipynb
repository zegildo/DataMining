## Extraindo informações do site da Globo

### Definindo variáveis globais necessárias

```python
URL = "https://www.globo.com/"
```

### Método que faz request do conteúdo da página

```python
def get_post_links(link):
    pagina = requests.get(link).content
    soup = BeautifulSoup(pagina)
    return soup.findAll('a', rel='bookmark')
    
```

### Método responsável por exibir os links e titulos das notícias da página

```python
def extract_data_from_link(post_link_tag):
    return{
        'link': post_link_tag.attrs['href'],
        'title': post_link_tag.text
            #o .attrs busca pelo parâmetro desejado, e ele retorna tudo que há dentro
    }
```

```python
posts = get_post_links(URL)
data = []
```

```python
for post in posts:
    post_data = extract_data_from_link(post)
    data.append(post_data)
```

## Parte 2 - Extraindo mensagens do site https://bitcointalk.org

### Declara as variáveis globais necessárias para o código funcionar

```python
cont = 0
last_page = 66960
value_page = 40 ##Representa o incremento numerico de cada página
```

### Cria o Método para realizar o request da página e atualizar o link quando necessário

```python

def update_link(val):
    URL2 = "https://bitcointalk.org/index.php?board=67."+str(val)
    page = requests.get(URL2)
    soup = BeautifulSoup(page.content, 'html.parser')
    tudo = soup.findAll('td', class_='windowbg3')
    tudo2 = soup.findAll('td', class_='windowbg')
    print("CONTEÚDOS DA PAGINA "+URL2)
    print_content(tudo, tudo2)
```

### Método para a exibição do conteúdo escolhido da página

```python
def print_content(content, content2):
    for post in content:
        print(post.text)

    for post2 in content2:
        print(post2.text)
```

### Laço que vai executar da primeira página até a ultima página do fórum

```python
while cont <= last_page:
    update_link(cont)
    cont+=value_page

```
