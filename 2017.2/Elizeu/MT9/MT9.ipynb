## Extraindo informaÃ§Ãµes do site da Globo

### Definindo variÃ¡veis globais necessÃ¡rias

```python
URL = "https://www.globo.com/"
```

### MÃ©todo que faz request do conteÃºdo da pÃ¡gina

```python
def get_post_links(link):
    pagina = requests.get(link).content
    soup = BeautifulSoup(pagina)
    return soup.findAll('a', rel='bookmark')
    
```

### MÃ©todo responsÃ¡vel por exibir os links e titulos das notÃ­cias da pÃ¡gina

```python
def extract_data_from_link(post_link_tag):
    return{
        'link': post_link_tag.attrs['href'],
        'title': post_link_tag.text
            #o .attrs busca pelo parÃ¢metro desejado, e ele retorna tudo que hÃ¡ dentro
    }
```

```python
posts = get_post_links(URL)
data = []
```

```python
for post in posts:
    post_data = extract_data_from_link(post)
    data.append(post_data)
```

## Parte 2 - Extraindo mensagens do site https://bitcointalk.org

### Declara as variÃ¡veis globais necessÃ¡rias para o cÃ³digo funcionar

```python
cont = 0
last_page = 66960
value_page = 40 ##Representa o incremento numerico de cada pÃ¡gina
```

### Cria o MÃ©todo para realizar o request da pÃ¡gina e atualizar o link quando necessÃ¡rio

```python

def update_link(val):
    URL2 = "https://bitcointalk.org/index.php?board=67."+str(val)
    page = requests.get(URL2)
    soup = BeautifulSoup(page.content, 'html.parser')
    tudo = soup.findAll('td', class_='windowbg3')
    tudo2 = soup.findAll('td', class_='windowbg')
    print("CONTEÃšDOS DA PAGINA "+URL2)
    print_content(tudo, tudo2)
```

### MÃ©todo para a exibiÃ§Ã£o do conteÃºdo escolhido da pÃ¡gina

```python
def print_content(content, content2):
    for post in content:
        print(post.text)

    for post2 in content2:
        print(post2.text)
```

### LaÃ§o que vai executar da primeira pÃ¡gina atÃ© a ultima pÃ¡gina do fÃ³rum

```python
while cont <= last_page:
    update_link(cont)
    cont+=value_page

```
