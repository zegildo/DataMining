```python
import math

#Dados
X = [6.6, 6.9, 7.3, 7.5, 8.2, 8.3, 9.1, 9.2, 9.4, 10.2]
Y = [86, 92, 71, 74, 185, 85, 201, 283, 255, 222]

def correlacao_Pearson(X, Y):
    media_x = sum(X)/len(X)
    media_y = sum(Y)/len(Y)

    #Numerador
    media_subtraida_x = [i - media_x for i in X]
    media_subtraida_y = [i - media_y for i in Y]

    x_vezes_y = [a * b for a, b in list(zip(media_subtraida_x, media_subtraida_y))]
    soma_x_vezes_y = sum(x_vezes_y)
    #Fim_Numerador
    
    #Denominador
    x_quadrado = [i * i for i in media_subtraida_x]
    y_quadrado = [i * i for i in media_subtraida_y]

    soma_x_quadrado = sum(x_quadrado)
    soma_y_quadrado = sum(y_quadrado)

    soma_x_quadrado_soma_y_quadrado = soma_x_quadrado*soma_y_quadrado
    raiz_soma_x_quadrado_soma_y_quadrado = math.sqrt(soma_x_quadrado_soma_y_quadrado)
    #Fim_Denominador
    
    #Resultado
    return soma_x_vezes_y/raiz_soma_x_quadrado_soma_y_quadrado

print ('Correlacao de Pearson (r): ', correlacao_Pearson(X, Y))
```

```python
from scipy import stats
import warnings 
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

warnings.filterwarnings('ignore') 
data_cp = {'Otolith_lenght': [6.6, 6.9, 7.3, 7.5, 8.2, 8.3, 9.1, 9.2, 9.4, 10.2], 
        'Fish_mass':    [86, 92, 71, 74, 185, 85, 201, 283, 255, 222]}
r = stats.pearsonr(data_cp['Otolith_lenght'], data_cp['Fish_mass'])
df = pd.DataFrame(data_cp)

sns.regplot(x="Otolith_lenght", y="Fish_mass", data=df);
plt.title('Gráfico de dispersão para a relação entre o comprimento do Otolith e a massa do peixe ');
plt.xlabel('Comprimentto do Otolith (mm)');
plt.ylabel('Massa do Peixe (g)');

print ('Relação entre o comprimento do Otolith e a massa do peixe: ', r)



```

# Conclusão

Como podemos ver, há uma relação linear forte entre a massa do peixe e o tamanho Otolith. O coeficiente de correlação r = 0,83 defende uma correlação linear positiva e o p<0.003 informa que é muito improvável que esse resultado seja obtido por acaso, indicando que existe uma relação significativa entre as variáveis testadas.
