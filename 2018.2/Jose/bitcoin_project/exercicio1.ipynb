```python
from lxml import html 
from lxml import etree 

def get_page(path):
    
    arquivo = open(path).read()
    pagina = etree.HTML(arquivo)
    
    return pagina

def get_card(pagina):

    subjects = []
    authors = []
    replies = []
    views = []
    last_post = []
    
    subject_wbg3 = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[@class="windowbg3"]/span/a/text()')
    subject_wbg = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[@class="windowbg"]/span/a/text()')
    
    authors_wbg2 = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[@class="windowbg2"]/a/text()')
    
    replies_wbg3 = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[5][@class="windowbg3" and @align="center"]/text()')
    replies_wbg = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[5][@class="windowbg" and @align="center"]/text()')
   
    views_wbg3 = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[6][@class="windowbg3" and @align="center"]/text()')
    views_wbg = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td[6][@class="windowbg" and @align="center"]/text()')
    
    #MAINTENANCE
    #last_post_wbg2 = pagina.xpath('///table[@class= "bordercolor"]/tbody/tr/td/span[@class="smalltext"]/text()/normalize-space(.)')
    
    subjects.extend(subject_wbg3)
    subjects.extend(subject_wbg)
    
    authors.extend(authors_wbg2)
    
    replies.extend(replies_wbg3)
    replies.extend(replies_wbg)
       
    views.extend(views_wbg3)
    views.extend(views_wbg)
    
   # last_post.extend(last_post_wbg2)
        
    return zip(subjects, authors, replies, views)

def print_inserts(card):
    print("Subjects, Started by, Replies, Views, Last Post(Maintenance)")
    for subjects, authors, replies, views in card:
         print(subjects +', ' + authors +', ' + replies.strip() +', ' + views.strip())
            
pagina = get_page('Altcoin Discussion.htm')
cards = get_card(pagina)
print_inserts(cards)    

```
