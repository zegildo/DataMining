### Lendo os dados BBC_Result

```python
import pandas as pd

df = pd.read_csv('BBC_Results.csv')
df
```

### Verifica o tipo de dados das colunas:

```python
df.dtypes
```

### Ajusta o tipo de dados das colunas:

```python
def ajusta_tipo_coluna(df, lista_colunas):
    df[lista_colunas] = df[lista_colunas].apply(pd.to_numeric, errors='coerce')

ajusta_tipo_coluna(df,['Recognition Score','Temporal Memory Score'])
```

```python
df
```

### Substitui valores nulos pela média:

```python
colunas = ['How many hours did you sleep last night?','Recognition Score','Temporal Memory Score']
for coluna in colunas:
    df[coluna].fillna(df[coluna].mean(), inplace=True)
df
```

### Substitui valores fora do limite pela média:

```python
colunas = ['Recognition Score','Temporal Memory Score']
for coluna in colunas:
    df[(df[coluna] > 100) | (df[coluna] < 0)] = df[coluna].mean()
df
```

### Salva o arquivo limpo .csv:

```python
df.to_csv('BBC_Results_final.csv', sep=',', encoding='utf-8', index=False)
print "Done!"
```
